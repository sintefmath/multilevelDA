### Ranks

```python
source_path = "/home/florianb/havvarsel/multilevelDA/scripts/RankHistograms/BasinSL/2023-06-05T09_34_39"
```

```python
def plot_histogram(T, mode, bins_size=1/25):
    # Read from file
    SL_ranks = np.load(source_path+"/SL"+mode+"Ranks_"+str(T)+".npy")
    
    fig, axs = plt.subplots(1, 3, figsize=(15,3))
    fig.suptitle(r"$\bf{SL ranks}$ for "+mode+" at "+str(T),y=1.1)

    axs[0].set_title("eta")
    axs[1].set_title("hu")
    axs[2].set_title("hv")

    bins = np.arange(0,1+1e-9,bins_size)

    for i in range(3):
        hist = np.histogram(SL_ranks[:,i], bins=bins)
        axs[i].bar(0.5*(hist[1][1:]+hist[1][:-1]),hist[0]/len(SL_ranks)/bins_size, width=bins_size, color="C0")
        
        axs[i].set_ylim((0,1.5))
        axs[i].set_xlim((0,1.0))

        axs[i].plot([0,1],[1,1], color="black", ls="--", lw=3)
```
