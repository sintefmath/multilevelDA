```python
def relu(X, loc=0, scale=1):
    """
    Rectified linear unit: 
    - Obtains values between 0 and 1.
    - Centered around loc and has slope 1/scale. 

    Input:
    (Using same syntax and output format as scipy.stats.norm.cdf)
    X     - ndarray of size(m,n)
    loc   - ndarray of size (n) 
    scale - ndarray of size(n) 
    """
    return np.minimum(np.maximum(0.0, (X.T-loc[:,np.newaxis])/scale[:,np.newaxis] +0.5), 1.0).T
```

```python
for n in range(N_ranks):
    print(n)
    ## New Ensemble
    # 0-level
    for e in range(ML_Nes[0]):
        MLOceanEnsemble.ML_ensemble[0][e].upload(data_args_list[0]["eta"], data_args_list[0]["hu"], data_args_list[0]["hv"])
        MLOceanEnsemble.ML_ensemble[0][e].t = 0.0
        init_mekls[0].perturbSim(MLOceanEnsemble.ML_ensemble[0][e])

    # diff-levels
    for l_idx in range(1,len(ML_Nes)):
        for e in range(ML_Nes[l_idx]):
            MLOceanEnsemble.ML_ensemble[l_idx][0][e].upload(data_args_list[l_idx]["eta"], data_args_list[l_idx]["hu"], data_args_list[l_idx]["hv"])
            MLOceanEnsemble.ML_ensemble[l_idx][0][e].t = 0.0
            
            MLOceanEnsemble.ML_ensemble[l_idx][1][e].upload(data_args_list[l_idx-1]["eta"], data_args_list[l_idx-1]["hu"], data_args_list[l_idx-1]["hv"])
            MLOceanEnsemble.ML_ensemble[l_idx][1][e].t = 0.0

            init_mekls[l_idx].perturbSim(MLOceanEnsemble.ML_ensemble[l_idx][0][e])
            init_mekls[l_idx-1].perturbSimSimilarAs(MLOceanEnsemble.ML_ensemble[l_idx][1][e], modelError=init_mekls[l_idx])

            # partner_mekls[l_idx].perturbSim(MLOceanEnsemble.ML_ensemble[l_idx][0][e])
            # partner_mekls[l_idx-1].perturbSim(MLOceanEnsemble.ML_ensemble[l_idx][1][e])

    ## New Truth
    truth.upload(data_args_list[-1]["eta"], data_args_list[-1]["hu"], data_args_list[-1]["hv"])
    truth.t = 0.0
    init_mekl.perturbSim(truth)

    ## Ranks
    ML_state = MLOceanEnsemble.download()
    true_state = np.array(truth.download(interior_domain_only=True))
    for h, [Hx, Hy] in enumerate(zip(rankHxs, rankHys)):
        lvlHx = int(Hx/(2**(len(ls)-1)))
        lvlHy = int(Hy/(2**(len(ls)-1)))

        lvl_ranks[0][n*len(rankHxs)+h] = 1/ML_Nes[0]*np.sum(1-relu(ML_state[0][:,lvlHy,lvlHx,:].T, loc=true_state[:,Hy,Hx], scale=R), axis=0)

        ML_ranks[n*len(rankHxs)+h] += lvl_ranks[0][n*len(rankHxs)+h]
        
        for l_idx in range(1, len(ls)):
            lvlHx = int(Hx/(2**(len(ls)-1-l_idx)))
            lvlHy = int(Hy/(2**(len(ls)-1-l_idx)))

            lvl_ranks[l_idx][n*len(rankHxs)+h] = 1/ML_Nes[l_idx] * np.sum(1-relu(ML_state[l_idx][0][:,lvlHy,lvlHx,:].T, loc=true_state[:,Hy,Hx], scale=R), axis=0)

            diff_lvl_ranks[l_idx-1][n*len(rankHxs)+h] = 1/ML_Nes[l_idx] * np.sum((1-relu(ML_state[l_idx][0][:,lvlHy,lvlHx,:].T, loc=true_state[:,Hy,Hx], scale=R))
                                                                                - (1-relu(ML_state[l_idx][1][:,int(lvlHy/2),int(lvlHx/2),:].T, loc=true_state[:,Hy,Hx], scale=R)), axis=0)

            ML_ranks[n*len(rankHxs)+h] += diff_lvl_ranks[l_idx-1][n*len(rankHxs)+h]
        
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15,3))
fig.suptitle(r"$\bf{ML ranks}$",y=1.1)

bins_size=1/25

axs[0].set_title("eta")
axs[1].set_title("hu")
axs[2].set_title("hv")

bins = np.arange(-0.5,1.5+1e-9,bins_size)

for i in range(3):
    hist = np.histogram(ML_ranks[:,i], bins=bins)
    axs[i].bar(0.5*(hist[1][1:]+hist[1][:-1]),hist[0]/len(ML_ranks)/bins_size, width=bins_size, color="C0")
    
    axs[i].set_ylim((0,1.5))
    # axs[i].set_xlim((0,1.0))

    axs[i].plot([0,1],[1,1], color="black", ls="--", lw=3)
```
