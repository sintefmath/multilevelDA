### Results

```python
workSL = np.loadtxt("workSL")
workML = np.loadtxt("workML")
errSL  = np.loadtxt("errSL")
errML  = np.loadtxt("errML")
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))

for a in range(3):
    axs[a].loglog(workSL, errSL[:,a], label="SL")
    axs[a].loglog(workML, errML[:,a], label="ML")
    axs[a].legend()
    axs[a].set_xlabel("theoretical work", fontsize=15)
    axs[a].set_ylabel("numerical error", fontsize=15)

axs[0].set_title("$\eta$", fontsize=15)
axs[1].set_title("$hu$", fontsize=15)
axs[2].set_title("$hv$", fontsize=15)
```
